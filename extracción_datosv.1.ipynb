```python
## - Explicar como usar la libreria spotipy 
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
# Coloca tus credenciales aquí
CLIENT_ID = ''
CLIENT_SECRET = ''

# Autenticación
auth_manager = SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python

#api original de spotipy

sp.playlist(f"https://open.spotify.com/playlist/0IyJJbmF3vp5nX6j4FFeDO?si=7aaea780041a4b63", fields=None, market=None, additional_types=('track',))
```

```python

# modificando Fields para extraer información de artista y españa
sp.playlist(f"https://open.spotify.com/playlist/0IyJJbmF3vp5nX6j4FFeDO?si=7aaea780041a4b63", fields="tracks.items(track(name,artists(name)))", market="ES" , additional_types=('track',))
```

```python

# agregando ID de artista para extraer género del artista

sp.playlist(f"https://open.spotify.com/playlist/0IyJJbmF3vp5nX6j4FFeDO?si=7aaea780041a4b63", fields="tracks.items(track(name,artists(id, name)))", market="ES" , additional_types=('track',))
```

```python
#itero para el genero de cada artista 

playlist = sp.playlist(playlist_id="0IyJJbmF3vp5nX6j4FFeDO",fields="tracks.items(track(name, artists(id, name)))")

for item in playlist["tracks"]["items"]:
    track = item["track"]
    artist = track["artists"][0]  # Primer artista de la canción
    artist_id = artist["id"]
    artist_name = artist["name"]

    # Obtener los géneros del artista
    artist_info = sp.artist(artist_id)
    genres = artist_info["genres"]

    print(f"Canción: {track['name']} - Artista: {artist_name} - Géneros: {', '.join(genres)}")
```
