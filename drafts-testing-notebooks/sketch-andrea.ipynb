```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
CLIENT_ID = 'introducir'
CLIENT_SECRET = 'introducir'

# Autenticaci√≥n
auth_manager = SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
info_raw = sp.playlist("4yofs4YkUFrTbfEJyQdzCW", fields=None, market=None, additional_types='track',)
```

```python
class Eurovision:

    def __init__(self,raw):
        self.raw = raw
        
    def a√±adir_datos(self):
        eurovision= []

        for item in self.raw["tracks"]["items"]:
            track = item["track"]
            album = item["track"]["album"]
            artist = item["track"]["artists"]
            
            eurovision.append({
                # Info canci√≥n    
                "track_name": track["name"],
                "track_id": track["id"],
                "popularity": track["popularity"],

                # Info artista    
                "artists_name": artist[0]["name"],
                "artists_id": artist[0]["id"],
                "artists_href": artist[0]["href"],

                # Info album  
                "album_name": album["name"],
                "id_album": album["id"],
                "album_type": album["album_type"],
                "release_date": album["release_date"],
                
                # Info playlist    
                "playlist_name": self.raw["name"],
                "playlist_followers": self.raw["followers"]["total"],
                "playlist_id": self.raw["id"],
                "playlist_href": self.raw["href"],
                "playlist_uri": self.raw["uri"],
                "playlist_url": self.raw["external_urls"]["spotify"]

            })

        
        df_eurovision = pd.DataFrame(eurovision)

        return df_eurovision


```

```python
eurovision_2021 = Eurovision(info_raw)
```
